```python
import copy
import timeit
import numpy as np
import matplotlib.pyplot as pl
%matplotlib inline
from ipywidgets import interactive
import ipywidgets as widgets
```

Create the following grid world.

**States:** A 10x10 grid

**Actions:** Up, Down, Left, Right

**Tranistion probabilities:**
* 0.7 in the direction of action
* 0.1 in the three other directions
* The robot bounces back to the same state near edges

**Rewards:**
* (7,8) has a reward +10
* (2,7) has a reward +3
* (4,3) has a reward -5
* (7,3) has a reward -10
* No reward in other states

This example is based on Decision Making Under Uncertainty by M.J. Kochenderfer.

```python
#Let's define MDP paras
def createGrid10World():
    def xy2s(y, x):
        x = max(x, 0)
        y = max(y, 0)
        x = min(x, 9)
        y = min(y, 9)
        out = np.ravel_multi_index(np.array([x,y]), (10,10))
        return out

    def s2xy(s):
        x, y = np.unravel_index(s, (10,10))
        return y, x

    def gridPlot(ax, im, title='', cmap='Blues'):
        pl.imshow(im, interpolation='none', cmap=cmap, origin='lower')
        pl.colorbar()
        ax.set_xticks(np.arange(0, 10, 1));
        ax.set_yticks(np.arange(0, 10, 1));
        ax.set_xticklabels(np.arange(0, 10, 1));
        ax.set_yticklabels(np.arange(0, 10, 1));
        ax.set_xticks(np.arange(-.5, 10, 1), minor=True);
        ax.set_yticks(np.arange(-.5, 10, 1), minor=True);
        ax.grid(which='minor', color='w', linestyle='-', linewidth=1)
        pl.title(title);
        return

    A = ['left', 'right', 'up', 'down']
    S = np.arange(100)
    T = np.zeros((len(S), len(A), len(S)))
    R = np.zeros((len(S), len(A)))
    for s in S:
        x, y = s2xy(s)
        if x == 2 and y == 7:
            R[s, :] = 3
        elif x == 7 and y == 8:
            R[s, :] = 10
        else:
            if x == 7 and y == 3:
                R[s, :] = -10
            elif x == 4 and y == 3:
                R[s, :] = -5
            elif x == 0:
                if y == 0 or y == 9:
                    R[s, :] = -0.2
                else:
                    R[s, :] = -0.1
                R[s, 0] = -0.7
            elif x == 9:
                if y == 0 or y == 9:
                    R[s, :] = -0.2
                else:
                    R[s, :] = -0.1
                R[s, 1] = -0.7
            elif y == 0:
                if x == 0 or x == 9:
                    R[s, :] = -0.2
                else:
                    R[s, :] = -0.1
                R[s, 3] = -0.7
            elif y == 9:
                if x == 0 or x == 9:
                    R[s, :] = -0.2
                else:
                    R[s, :] = -0.1
                R[s, 2] = -0.7

            for a in A:
                if a == 'left':
                    T[s, 0, xy2s(x - 1, y)] += 0.7
                    T[s, 0, xy2s(x + 1, y)] += 0.1
                    T[s, 0, xy2s(x, y - 1)] += 0.1
                    T[s, 0, xy2s(x, y + 1)] += 0.1
                elif a == 'right':
                    T[s, 1, xy2s(x + 1, y)] += 0.7
                    T[s, 1, xy2s(x - 1, y)] += 0.1
                    T[s, 1, xy2s(x, y - 1)] += 0.1
                    T[s, 1, xy2s(x, y + 1)] += 0.1
                elif a == 'up':
                    T[s, 2, xy2s(x, y + 1)] += 0.7
                    T[s, 2, xy2s(x, y - 1)] += 0.1
                    T[s, 2, xy2s(x - 1, y)] += 0.1
                    T[s, 2, xy2s(x + 1, y)] += 0.1
                elif a == 'down':
                    T[s, 3, xy2s(x, y - 1)] += 0.7
                    T[s, 3, xy2s(x, y + 1)] += 0.1
                    T[s, 3, xy2s(x - 1, y)] += 0.1
                    T[s, 3, xy2s(x + 1, y)] += 0.1

    for a, c_x, c_y in [(0,0,0), (0,0,9),(1,9,0),(1,9,9),(2,0,9),(2,9,9),(3,0,0),(3,9,0)]:
        R[xy2s(c_x,c_y),a] = -0.8

    discount = 0.9

    nextStates = {}
    for si in range(len(S)):
        for ai in range(len(A)):
            nextStates[(si,ai)] = np.where((T[si, ai, :] != 0) == True)[0]

    return {'S':S, 'A':A, 'T':T, 'R':R, 'discount':discount, 'nextStates':nextStates, 'gridPlot':gridPlot, 'xy2s':xy2s, 's2xy':s2xy}
```

```python
class MDP():
    def __init__(self):
        pass

    def inbuilt_init(self, mdp_dict):
        self.S = mdp_dict['S']
        self.A = mdp_dict['A']
        self.T = mdp_dict['T']
        self.R = mdp_dict['R']
        self.discount = mdp_dict['discount']
        self.nextStates = mdp_dict['nextStates']
        self.gridPlot = mdp_dict['gridPlot']
        self.xy2s = mdp_dict['xy2s']
        self.s2xy = mdp_dict['s2xy']
```

```python
#Define the MDP
mdp = MDP()
mdp.inbuilt_init(mdp_dict=createGrid10World())

#Plot states
pl.figure(figsize=(3,3))
mdp.gridPlot(ax=pl.gca(), im=mdp.S.reshape((10,10)), title='States', cmap='Greys')

#Plot rewards
pl.figure(figsize=(15,3))
pl.suptitle('Rewards', fontsize=15)
for a in range(4):
    pl.subplot(1,4,a+1)
    mdp.gridPlot(ax=pl.gca(), im=mdp.R[:,a].reshape((10,10)), title='a='+mdp.A[a], cmap='jet')
pl.show()

#Plot rewards
pl.figure(figsize=(15,3))
pl.suptitle('Rewards - clipped larger +ve values for visualizing the edges', fontsize=15)
for a in range(4):
    pl.subplot(1,4,a+1)
    mdp.gridPlot(ax=pl.gca(), im=np.clip(mdp.R[:,a].reshape((10,10)), -1, -0.5), title='a='+mdp.A[a], cmap='jet')
pl.show()

#Plot rewards
s0_x, s0_y = 3, 5
s0 = mdp.xy2s(s0_x, s0_y)
pl.figure(figsize=(15,3))
pl.suptitle('Transition probabilities T(s1|s0=({},{}),a)'.format(s0_x, s0_y), fontsize=15)
for a in range(4):
    pl.subplot(1,4,a+1)
    mdp.gridPlot(ax=pl.gca(), im=mdp.T[s0,a,:].reshape((10,10)), title='a='+mdp.A[a], cmap='Blues')
pl.show()
```

### 1. Policy evaluation

Computing the utility, U.

$U^\pi_k(s) = R(s, \pi(s)) + \gamma \sum_{s'} T(s' \mid s, \pi(s))U^\pi_{k-1}(s')$

```python
#edited....
def iterativePolicyEvaluation(mdp, policy, numIterations=10):
    U = np.zeros(len(mdp.S))
    for _ in range(numIterations):
      #type your code here
        U_old = U.copy()
        for s in range(len(mdp.S)):
            a = policy if np.isscalar(policy) else policy[s]
            U[s] = np.sum(mdp.T[s, a, :] * (mdp.R[s, a,] + mdp.discount * U_old))  # or the per-transition version
    return U

numIterations = 5
pl.figure(figsize=(15,3))
pl.suptitle('Utilities', fontsize=15)
for a in range(4):
    pl.subplot(1,4,a+1)
    U = iterativePolicyEvaluation(mdp=mdp, policy=a, numIterations=numIterations)
    mdp.gridPlot(ax=pl.gca(), im=U.reshape(10,10), title='a='+mdp.A[a], cmap='jet')
pl.show()
#print(np.round(U.reshape(10,10),3))
```

### 2. Policy iteration

Policy evaluation can be used in policy iteration:
1. Given the current policy, compute U
2. Using U, compute a new policy

```python
def policyIteration(mdp, numIterations=1):
    U_pi_k = np.zeros(len(mdp.S))                                  # initial values
    pi_k   = np.random.randint(low=0, high=4, size=len(mdp.S), dtype=int)  # initial policy
    pi_kp1 = copy.copy(pi_k)
#edited
    for t in range(numIterations):
        # Policy evaluation: compute U_pi_k for the current policy
         #type your code here
        U_pi_k = iterativePolicyEvaluation(mdp=mdp, policy=pi_k, numIterations=20)

        # Policy improvement: choose the best action in each state using U_pi_k
        #type your code here
        for s in range(len(mdp.S)):
            if mdp.R.ndim == 2:
                # R is (S, A)
                q_sa = mdp.R[s, :] + mdp.discount * (mdp.T[s, :, :] @ U_pi_k)   # shape (A,)
            else:
                # R is (S, A, S)
                q_sa = np.sum(mdp.T[s, :, :] * (mdp.R[s, :, :] + mdp.discount * U_pi_k), axis=1)
            pi_kp1[s] = int(np.argmax(q_sa))

        # next iteration uses the improved policy
        pi_k = pi_kp1.copy()

    return U_pi_k, pi_kp1

U_pi_k, pi_kp1 = policyIteration(mdp, numIterations=2)
```

```python
import time, numpy as np, matplotlib.pyplot as plt

def valueIteration_until_converged(mdp, theta=1e-6, max_iters=2000):
    U, deltas = np.zeros(len(mdp.S)), []
    t0 = time.perf_counter()
    for it in range(max_iters):
        U_old = U.copy()
        for s in range(len(mdp.S)):
            if mdp.R.ndim == 2:  # R(s,a)
                q_sa = mdp.R[s, :] + mdp.discount * (mdp.T[s, :, :] @ U_old)
            else:                # R(s,a,s')
                q_sa = np.sum(mdp.T[s, :, :] * (mdp.R[s, :, :] + mdp.discount * U_old), axis=1)
            U[s] = np.max(q_sa)
        delta = np.max(np.abs(U - U_old))
        deltas.append(delta)
        if delta < theta: break
    elapsed = time.perf_counter() - t0
    return U, (it+1), deltas, elapsed
```

```python
def _sample_step(mdp, s, a, rng):
    """Sample s' ~ T[s,a,*] robustly; handle terminals and empty next-state sets."""
    # If current state is terminal, stay there
    if hasattr(mdp, "terminal_states") and (s in mdp.terminal_states):
        r = mdp.R[s, a] if mdp.R.ndim == 2 else 0.0
        return s, r, True

    if hasattr(mdp, "nextStates"):
        idx = mdp.nextStates.get((s, a), [])
        # No valid next states  self-loop
        if len(idx) == 0:
            r = mdp.R[s, a] if mdp.R.ndim == 2 else 0.0
            done = hasattr(mdp, "terminal_states") and (s in mdp.terminal_states)
            return s, r, done

        p = mdp.T[s, a, idx].astype(float)
        p = np.clip(p, 0.0, None)
        psum = p.sum()
        p = (np.ones(len(idx)) / len(idx)) if psum <= 0 else (p / psum)
        s1 = int(rng.choice(idx, p=p))
    else:
        # Full distribution path
        p = mdp.T[s, a, :].astype(float)
        p = np.clip(p, 0.0, None)
        psum = p.sum()
        if psum <= 0:
            s1 = s  # self-loop
        else:
            s1 = int(rng.choice(np.arange(len(mdp.S)), p=p / psum))

    r = mdp.R[s, a] if mdp.R.ndim == 2 else mdp.R[s, a, s1]
    done = hasattr(mdp, "terminal_states") and (s1 in mdp.terminal_states)
    return s1, r, done



def q_learning_on_mdp(mdp, episodes=4000, max_steps=500, alpha=0.1, epsilon=0.1, tol=1e-4):
    A = mdp.T.shape[1]; Q = np.zeros((len(mdp.S), A))
    rewards, rng = [], np.random.default_rng()
    gamma, t0, ep_tol = mdp.discount, time.perf_counter(), None

    for ep in range(episodes):
        s = int(rng.integers(0, len(mdp.S)))
        if hasattr(mdp, "terminal_states"):
            while s in mdp.terminal_states: s = int(rng.integers(0, len(mdp.S)))
        ep_ret, maxchg = 0.0, 0.0
        for _ in range(max_steps):
            # -greedy
            if rng.random() < epsilon: a = int(rng.integers(0, A))
            else:
                m = np.max(Q[s]); a = int(rng.choice(np.where(Q[s]==m)[0]))
            s1, r, done = _sample_step(mdp, s, a, rng)
            old = Q[s, a]
            Q[s, a] = old + alpha * (r + gamma * np.max(Q[s1]) - old)
            maxchg = max(maxchg, abs(Q[s, a] - old))
            ep_ret += r; s = s1
            if done: break
        rewards.append(ep_ret)
        if ep_tol is None and maxchg < tol: ep_tol = ep+1

    elapsed = time.perf_counter() - t0
    if ep_tol is None: ep_tol = episodes
    policy = np.argmax(Q, axis=1).astype(int)
    return Q, policy, rewards, ep_tol, elapsed
```

```python
def moving_avg(x, k=50):
    x = np.asarray(x);
    return np.convolve(x, np.ones(k)/k, mode='valid') if len(x)>=k else x

# Model-Based
U_mb, it_mb, deltas_mb, time_mb = valueIteration_until_converged(mdp, theta=1e-6)
print(f"MODEL-BASED (Value Iteration): time={time_mb:.3f}s, iters={it_mb}, final delta={deltas_mb[-1]:.2e}")

# Model-Free
Q, pi_mf, rewards, eps_tol, time_mf = q_learning_on_mdp(mdp, episodes=4000, tol=1e-4)
print(f"MODEL-FREE (Q-Learning):  time={time_mf:.3f}s, episodes_to_tol={eps_tol}")

# Convergence plots
plt.figure(); plt.plot(deltas_mb); plt.title("Model-Based: max|U| per iteration"); plt.xlabel("Iteration"); plt.ylabel("max|U|"); plt.show()
plt.figure(); plt.plot(rewards, alpha=.4); plt.plot(moving_avg(rewards,100));
plt.title("Model-Free: return per episode (with MA100)"); plt.xlabel("Episode"); plt.ylabel("Return"); plt.show()
```
